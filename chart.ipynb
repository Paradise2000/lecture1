```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Salary Data (2).csv")

avg_salary = df.groupby("Gender")["Salary"].mean()

plt.figure(figsize=(6,4))
avg_salary.plot(kind='bar', color=['blue', 'red'])
plt.xlabel("Płeć")
plt.ylabel("Średnia pensja (USD)")
plt.title("Średnie wynagrodzenie według płci")

plt.show()
```
